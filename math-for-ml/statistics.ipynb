### 1. PDF and KDE

```python
# 1. using seaborn plot
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

#generate a random normal distribution
np.random.seed(42)
data = np.random.normal(loc=0, scale=1, size=1000)

# plotting
fig, axis = plt.subplots(1,3, figsize=(20,5))
ax1, ax2, ax3 = axis

#plotting kde-plot from seaborn
sns.kdeplot(data, bw_adjust=0.5, fill=True, label='Seaborn KDE', color='Orange', ax=ax1)
plt.title('Probability Density Function (PDF) using KDE')
plt.xlabel('Data')
plt.ylabel('Density')

sns.kdeplot(data, bw_adjust=0.75, fill=False, label='Seaborn KDE', color='Blue', ax=ax2)
plt.title('Probability Density Function (PDF) using KDE')
plt.xlabel('Data')
plt.ylabel('Density')

sns.kdeplot(data, bw_adjust=1, fill=True, color='Red', label='Seaborn KDE', ax=ax3)
plt.title('Probability Density Function (PDF) using KDE')
plt.xlabel('Data')
plt.ylabel('Density')

plt.show()
```

```python
### 2. same pdf using scipy kde
from scipy.stats import gaussian_kde

# generate data
data = np.random.normal(loc=0, scale=1, size=1000)

#kde using scipy
kde_scipy = gaussian_kde(data)
x = np.linspace(min(data), max(data), 1000)
pdf_scipy = kde_scipy(x)

#plotting
plt.figure(figsize=(5,5))
plt.plot(x, pdf_scipy, label='Scipy KDE', color='blue')
plt.show()
```
