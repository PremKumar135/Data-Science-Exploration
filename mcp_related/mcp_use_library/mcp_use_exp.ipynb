```python
from langchain_openai import ChatOpenAI
llm = ChatOpenAI(model='gpt-4o-mini')
```

### 1.1 Single config

```python
import nest_asyncio
nest_asyncio.apply()

import asyncio
import os
from mcp_use import MCPAgent, MCPClient
from textwrap import dedent

airbnb_config = {
        "mcpServers": {
            "airbnb": {
                "command": "npx",
                "args": ["-y", "@openbnb/mcp-server-airbnb", "--ignore-robots-txt"]
            }
        }
    }


async def run_airbnb_example(query):
    client = MCPClient.from_dict(airbnb_config)
    agent = MCPAgent(llm=llm, client=client, max_steps=30)

    try:
        result = await agent.run(query, max_steps=30,)
        print(f"Result: {result}")

    finally:
        if client.sessions:
            await client.close_all_sessions()


if __name__ == "__main__":
    question = dedent("""
            Find me a nice place to stay in Chennai for 2 adults 
            for a week in August. I prefer places with a pool and 
            good reviews. Show me the top 3 options.
        """)
    asyncio.run(run_airbnb_example(question))
```

### 3. Using Langchain Adapter with mcp-use library

```python
import asyncio
from langchain_openai import ChatOpenAI
from langgraph.prebuilt import create_react_agent
from mcp_use.client import MCPClient
from mcp_use.adapters.langchain_adapter import LangChainAdapter
from dotenv import load_dotenv

load_dotenv()


async def main():
    # Initialize MCP client
    client = MCPClient.from_config_file("./airbnb_config.json")
    llm = ChatOpenAI(model="gpt-4o-mini")

    # Create adapter instance and Get LangChain tools with a single line
    adapter = LangChainAdapter()
    tools = await adapter.create_tools(client)
    print(tools)

    # Create a custom LangChain agent
    agent = create_react_agent(model=llm, tools=tools)
    result = await agent.ainvoke( {"messages": [{"role": "user", 
                                                "content": "What tools do you have avilable ? "
                                                }]
                                })
    print(result)


if __name__ == "__main__":
    asyncio.run(main())
```
