### 1. Use client(mcp-use) to connect to fastapi mcp servers
- step1: run main.py as mcp server
- step2 : run below cell to access the fastapi endpoints as mcp servers

```python
import nest_asyncio
nest_asyncio.apply()
import asyncio
from dotenv import load_dotenv
load_dotenv()
from mcp_use import MCPAgent, MCPClient
from langchain_openai import ChatOpenAI

llm = ChatOpenAI()


async def run_multi_server_example():
    client = MCPClient.from_config_file('./config.json')
    agent = MCPAgent(llm=llm, client=client, max_steps=30, memory_enabled=True)

    print('========== Interactive MCP Chat ==========')
    print('Type exit or quit to end the conversation')
    print('Type clear to clear conversation history')
    print('========================================')

    try:
        while True:
            user_input = input('\nYou: ')
            if user_input.lower() in ['exit', 'quit']:
                print('Ending conversation')
                break

            elif user_input.lower()=='clear':
                agent.clear_conversation_history()

            else:
                result = await agent.run(user_input, max_steps=30)
                print(result)
    finally:
        # Ensure we clean up resources properly
        if client.sessions:
            await client.close_all_sessions()

if __name__ == "__main__":
    asyncio.run(run_multi_server_example())
```
